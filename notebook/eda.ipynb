```python
import pandas as pd
```

```python
file_path = "../data/ModifiedScatsDataOctober2006 .csv"

# Load without header
df_raw = pd.read_csv(file_path, header=None)

# Combine first 2 rows properly
new_columns = []

for time_label, col_name in zip(df_raw.iloc[0], df_raw.iloc[1]):
    time_label = str(time_label).strip()
    col_name = str(col_name).strip()
    
    if col_name.startswith("V"):
        new_columns.append(f"{col_name}_{time_label}")
    else:
        new_columns.append(col_name)

# Remove first 2 rows
df = df_raw.iloc[2:].copy()

# Apply fixed headers
df.columns = new_columns

# Reset index
df.reset_index(drop=True, inplace=True)

df.head()
```

saving new data

```python
# Save cleaned dataset
df.to_csv("../data/cleaned_scats_data.csv", index=False)

print("Cleaned dataset saved successfully!")
```

Identify traffic columns

```python
traffic_cols = [
    col for col in df_new.columns
    if col.startswith("V")
]

print(traffic_cols[:5])
print("Total traffic columns:", len(traffic_cols))
```

```python
df_long = df_new.melt(
    id_vars=["SCATS Number", "Date"],
    value_vars=traffic_cols,
    var_name="TimeSlot",
    value_name="Traffic"
)
df_long.to_csv("../data/scats_time_series.csv", index=False)

print("Time-series dataset saved!")

df_long.head()
```

Creating individual time column

```python
df_long[["TimeSlot", "Time"]] = df_long["TimeSlot"].str.split("_", expand=True)

df_long.head()
```

```python
#coneecting the time slot and time into one column for better analysis
df_long["Datetime"] = pd.to_datetime(
    df_long["Date"] + " " + df_long["Time"],
    dayfirst=True
)

df_long.head()
```

```python
df_long[df_long["Time"] != "0:00"].head() #verifiying the time slot and time split is correct
```

```python
#sorting data cronologically as Lstm model is sensitive to the order of data

#doing this to test and practice evwn thou the data seems to be in orer already
df_long = df_long.sort_values(
    by=["SCATS Number", "Datetime"]
)

df_long.head()
```

Check duplicate timestamps

```python
duplicates = df_long.groupby(
    ["SCATS Number", "Datetime"]
).size()

duplicates[duplicates > 1].head(10)
```

Aggregate duplicate timestamps

###LSTM/GRU forecasting models expect 1 value for each time slot..so we will mean it###


```python
df_long = df_long.groupby(
    ["SCATS Number", "Datetime"],
    as_index=False
)["Traffic"].mean()

df_long.head()
```

```python
df_long.shape
```

```python
df_long.isnull().sum()
```

Date and days feature

```python
df_long["Hour"] = df_long["Datetime"].dt.hour

df_long.head()
```

```python
df_long["DayOfWeek"] = df_long["Datetime"].dt.dayofweek

df_long.head()
```

Feature engineering

```python
#Weekend Feature
df_long["IsWeekend"] = df_long["DayOfWeek"].isin([5, 6]).astype(int)
df_long.head()
```

Create Lag 1 Feature

```python
df_long["Lag1"] = df_long.groupby(
    "SCATS Number"
)["Traffic"].shift(1)

df_long.head(10)
```
