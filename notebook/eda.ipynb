```python
import pandas as pd
```

```python
file_path = "../data/ModifiedScatsDataOctober2006 .csv"

# Load without header
df_raw = pd.read_csv(file_path, header=None)

# Combine first 2 rows properly
new_columns = []

for time_label, col_name in zip(df_raw.iloc[0], df_raw.iloc[1]):
    time_label = str(time_label).strip()
    col_name = str(col_name).strip()
    
    if col_name.startswith("V"):
        new_columns.append(f"{col_name}_{time_label}")
    else:
        new_columns.append(col_name)

# Remove first 2 rows
df = df_raw.iloc[2:].copy()

# Apply fixed headers
df.columns = new_columns

# Reset index
df.reset_index(drop=True, inplace=True)

df.head()
```

saving new data

```python
# Save cleaned dataset
df.to_csv("../data/cleaned_scats_data.csv", index=False)

print("Cleaned dataset saved successfully!")
```

```python


df_new = df_new.loc[:, ~df_new.columns.str.contains("nan", case=False)]

df_new.shape
```

Identify traffic columns

```python
traffic_cols = [
    col for col in df_new.columns
    if col.startswith("V")
]

print(traffic_cols[:5])
print("Total traffic columns:", len(traffic_cols))
```

```python
df_long = df_new.melt(
    id_vars=["SCATS Number", "Date"],
    value_vars=traffic_cols,
    var_name="TimeSlot",
    value_name="Traffic"
)
df_long.to_csv("../data/scats_time_series.csv", index=False)

print("Time-series dataset saved!")

df_long.head()
```

Creating individual time column

```python
df_long[["TimeSlot", "Time"]] = df_long["TimeSlot"].str.split("_", expand=True)

df_long.head()
```
